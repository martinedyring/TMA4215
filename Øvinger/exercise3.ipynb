**Construct the input**

```python
def construct_A(n):
    """
    Function to construct the real n x n matrix A, 
    where the element are defined as
    a_ij = 3^(- abs(i-j)) + 2^(-i.j) + 10^(-6)*np.random.uniform(0, 1)

    Args:
        n (int): size of square matrix A
        
    Returns:
        np.ndarray: 2D array containing data with float type
    """
    def _element_a(i, j):
        return 3.**(-np.abs(i-j)) + 2.**(-i-j) + 1e-6 * np.random.uniform(0, 1)
    
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            # add 1 since the matrix index goes from 1 to n, not 0 to n-1
            A[i, j] = _element_a(i+1, j+1) 
    return A

def construct_B(n):
    """
    Function to construct the vector b of n elements, 
    where the elements are defines as
    b = (1., 1., ..., 1.)^T

    Args:
        n (int): size of vector b

    Returns:
        np.ndarray: 1D array containing data with float type
    """
    b = np.full(shape=(n), fill_value=1.)
    return b
```

**Richardson iteration**

```python
def richardson(A, b, x_0, w, maxiter, tol, verbose=False):
    """
    Richardson Iteration to solve Ax=b
    
    Note: This function will check whether the choice of 'w' give rise to a converge iteration method. 
    The function will rise an error if not.

    Args:
        A (np.ndarray): 2D array representing matrix A
        b (np.ndarray): 1D array representing b
        x_0 (np.array): 1D array representing initial start of iteration
        w (float): a sixth parameter to the function
        maxiter (int): maximun number of iteration
        tol (float): tolerance for stopping criterion
        verbose (bool, optional): Whether to print the relative error for each iteration. Defaults to False.

    Returns:
        tuple(np.ndarray, np.ndarray): 1D array of soluton x, 1D array of relative error for each iteration
    """
    # Chech whether the algorithm will converge
    I = np.identity(len(b))
    B = I - w*A
    assert np.abs(np.linalg.eigvals(B)).max() < 1, "The spectral radius if A is greater than 1, thus the algorithm will not converge."

    # Richardson iteration
    r_0 = b - A.dot(x_0)
    x_old = x_0

    error_values = []
    for k in range(maxiter):
        r_k = b - A.dot(x_old)
        x_new = x_old + w*r_k
        x_old = x_new
        
        error = np.linalg.norm(r_k) / np.linalg.norm(r_0)
        error_values.append(error)
        if verbose:
            print(f"Error= {error}")
        if error < tol and k>0:
            break

    return x_new, error_values
```

**Gauss-Seidel Method**

```python

def gaussseidel(A, b, x_0, maxiter, tol, verbose=False):
    """
    Gauss-Seidel Method to solve Ax=b
    
    Args:
        A (np.ndarray): 2D array representing matrix A
        b (np.ndarray): 1D array representing b
        x_0 (np.array): 1D array representing initial start of iteration
        maxiter (int): maximun number of iteration
        tol (float): tolerance for stopping criterion
        verbose (bool, optional): Whether to print the relative error for each iteration. Defaults to False.

    Returns:
        tuple(np.ndarray, np.ndarray): 1D array of soluton x, 1D array of relative error for each iteration
    """
    r_0 = b - A.dot(x_0)
    x_old = np.array(x_0, dtype=np.float64)
    error_values = []
    
    for k in range(maxiter):        
        x_new = np.zeros(shape=(len(x_old)), dtype=np.float64)
        for j in range(len(x_new)):
            sum1 = A[j, :j].dot(x_new[:j])
            sum2 = A[j, (j+1):].dot(x_old[(j+1):])
            x_new[j] = (1 / A[j, j]) * (b[j] - sum1 - sum2)
            
        r_k = b - A.dot(x_old)
        error = np.linalg.norm(r_k) / np.linalg.norm(r_0)
        x_old = x_new
        error_values.append(error)
        if verbose:
            print(f"Error= {error}")
        if error < tol:
            break

    return x_new, error_values
```

**Gradient Descent with "perfect" linesearch**

```python
def steepestdescent(A, b, x_0, maxiter, tol, verbose=False):
    """
    Gradient Descent with 'perfect' linesearch Iteration to solve Ax=b
    
    Args:
        A (np.ndarray): 2D array representing matrix A
        b (np.ndarray): 1D array representing b
        x_0 (np.array): 1D array representing initial start of iteration
        maxiter (int): maximun number of iteration
        tol (float): tolerance for stopping criterion
        verbose (bool, optional): Whether to print the relative error for each iteration. Defaults to False.

    Returns:
        tuple(np.ndarray, np.ndarray): 1D array of soluton x, 1D array of relative error for each iteration
    """
    r_0 = b - A.dot(x_0)

    r_new = r_0
    x_old = x_0
    error_values = []

    for k in range(maxiter):
        r_old = r_new
        w_k = (np.transpose(r_old).dot(r_old)) / (np.transpose(r_old).dot(A.dot(r_old)))
        x_new = x_old + w_k * r_old
        r_new = r_old - w_k * A.dot(r_old)
        
        error = np.linalg.norm(r_old) / np.linalg.norm(r_0)
        x_old = x_new
        error_values.append(error)
        if verbose:
            print(f"Error= {error}")
        if error < tol and k > 0:
            break
    return x_new, error_values
```

**Conjugate Gradient Iteration**

```python
def cg(A, b, x_0, maxiter, tol, verbose=False):
    """
    Conjugate Gradient Iteration to solve Ax=b
    
    Args:
        A (np.ndarray): 2D array representing matrix A
        b (np.ndarray): 1D array representing b
        x_0 (np.array): 1D array representing initial start of iteration
        maxiter (int): maximun number of iteration
        tol (float): tolerance for stopping criterion
        verbose (bool, optional): Whether to print the relative error for each iteration. Defaults to False.

    Returns:
        tuple(np.ndarray, np.ndarray): 1D array of soluton x, 1D array of relative error for each iteration
    """
    r_0 = b - A.dot(x_0)
    v_0 = r_0

    r_new = r_0
    v_new = v_0
    x_old = x_0
    error_values = []

    for k in range(maxiter):
        r_old = r_new
        v_old = v_new
        
        w_k = (np.transpose(r_old).dot(r_old)) / (np.transpose(v_old).dot(A.dot(v_old)))
        x_new = x_old + w_k * v_old
        r_new = r_old - w_k * A.dot(v_old)

        beta_k = (np.transpose(r_new).dot(r_new)) / (np.transpose(r_old).dot(r_old))
        v_new = r_new + beta_k * v_old
        
        error = np.linalg.norm(r_old) / np.linalg.norm(r_0)
        x_old = x_new
        error_values.append(error)
        if verbose:
            print(f"Error= {error}")
        if error < tol and k > 0:
            break
    return x_new, error_values
```
