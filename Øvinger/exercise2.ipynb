### **Problem 2:**

**a) Find the first three polynomials orthogonal to the inner product
$
\begin{equation}
\langle f, g \rangle = \int_0^1 xf(x)g(x) dx
\end{equation}
$**

We have that two polynomals $f$ and $g$ are orthogonal to the inner product if 
$
\begin{equation}
\langle f, g \rangle = \int_0^1 xf(x)g(x) dx = 0
\end{equation}
$

In order to find the three first polynomials that are orthogonal to the inner product, I start out with the three linearly independent polynomials of lowest order:
$ v_0 = 1, v_1 = x, v_2 = x^2 $.

I will then use Gram Schmidt to make $v_0, v_1, v_2$ orthogonal to the inner product.

Start with 
$
\begin{equation}
P_0(x) = v_0 = 1
\end{equation}
$

Want to find a projection of $v_1$ such that we have orthogonality with $P_0(x)$0 = \langle P_0, P_1\rangle_w$. Thus
$
\begin{align}
P_1(x) &= v_1 - \frac{\langle v_1, P_0\rangle_w}{\langle P_0, P_0\rangle_w} P_0 \\
&, \langle v_1, P_0\rangle_w = \langle x, 1\rangle_w = \int_0^1 x^2 dx = [\frac{1}{3}x^3]_0^1 = \frac{1}{3} \\
&, \langle P_0, P_0\rangle_w = \langle 1, 1\rangle_w = \int_0^1 x dx = [\frac{1}{2}x^2]_0^1 = \frac{1}{2} \\
\Rightarrow P_1(x) &= x - \frac{1/3}{1/2} * 1 = x- \frac{2}{3}
\end{align}
$

Next, want to find a projection of $v_2$ such that we have orthogonality with $P_0$ and $P_1$.

Use Gram Schmidt:
$
\begin{align}
P_2(x) &= v_2 - \frac{\langle v_2, P_1\rangle_w}{\langle P_1, P_1\rangle_w} P_1 - \frac{\langle v_2, P_0\rangle_w}{\langle P_0, P_0\rangle_w} P_0 \\
&, \langle v_2, P_1\rangle_w = \langle x^2, x-\frac{2}{3}\rangle_w = \int_0^1 x^4 - \frac{2}{3}x^3 dx = [\frac{1}{5}x^5 - \frac{2}{2}\frac{1}{4}x^4]_0^1 = \frac{1}{30} \\
&, \langle v_2, P_0\rangle_w = \langle x^2, 1\rangle_w = \int_0^1 x^3 dx = [\frac{1}{4}x^4]_0^1 = \frac{1}{4} \\
&, \langle P_1, P_1\rangle_w = \langle x-\frac{2}{3}, x-\frac{2}{3}\rangle_w = \int_0^1 x^3-2\frac{2}{3}x^2 + \left(\frac{2}{3}\right)^2x dx = [\frac{1}{4}x^4 - \frac{4}{3}\frac{1}{3}x^3 + \left(\frac{2}{3}\right)^2 \frac{1}{2}x^2]_0^1 = \frac{1}{36} \\
\Rightarrow P_2(x) &= x^2 - \frac{1/30}{1/36} * \left(x-\frac{2}{3}\right)- \frac{1/4}{1/2} * 1  = x^2 +\frac{6}{5} + \frac{3}{10}
\end{align}
$

In summary, the three first polynomials orthogonal to the inner product is
$\begin{equation}
P_0(x) = 1, P_1(x) = x-\frac{2}{3}, P_2(x) = x^2+\frac{6}{5}+\frac{3}{10}
\end{equation}$

**a) Implement the two definitions of the Chebyshev polynomials seen in class. Use them to plot $T_n(x)$ for $n$ up to $6, x\in [-1, 1]$ (in two different images) to check that the two definitions are equivalent**

```python
import numpy as np
import matplotlib.pyplot as plt

def chebyshev_T(n, x):
    return np.cos(n * np.arccos(x))

def chebyshev_method_1(num_n, num_x):
    """
        T_n(x) = cos(n * arccos(x))
    """
    n_values = np.linspace(0, num_n-1, num_n)
    x_values = np.linspace(-1, 1, num_x)

    T_n = np.zeros((num_n, num_x))
    for i_n in range(num_n):
        for i_x in range(num_x):
            T_n[i_n, i_x] = chebyshev_T(n=n_values[i_n], x=x_values[i_x])
            
    return T_n, x_values

num_x = 100
num_n = 7

T_1, x_values = chebyshev_method_1(num_n, num_x)


for i_n in range(num_n):
    plt.plot(x_values, T_1[i_n, :], label=f"n = {int(i_n)}")
plt.legend()
plt.title("Chebychev ploynomials created with method #1")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def chebyshev_method_2(num_n, num_x):
    """
        Recursive formula:
        T_0(x) = 1
        T_1(x) = x
        T_{n+1}(x) = 2xT_n(x)-T_{n-1}(x)
    """
    n_values = np.linspace(0, num_n-1, num_n)
    x_values = np.linspace(-1, 1, num_x)

    T_n = np.zeros((num_n, num_x))
    T_n[0, :] = 1
    T_n[1, :] = x_values
    for i_n in range(2, num_n):
        for i_x in range(num_x):
            T_n[i_n, i_x] = 2 * x_values[i_x] * T_n[i_n - 1, i_x] - T_n[i_n - 2, i_x]
            
    return T_n, x_values

num_x = 100
num_n = 7

T_2, x_values = chebyshev_method_2(num_n, num_x)


for i_n in range(num_n):
    plt.plot(x_values, T_2[i_n, :], label=f"n = {int(i_n)}")
plt.legend()
plt.title("Chebychev ploynomials created with method #2")
plt.show()
```

**b) 2) Implement a function "cheb_nodes(N)" which computes the Chebyshev nodes for a given $N$.**

The Chebyshev nodes are roots for the Chebyshev polynomial. We plug the intervall $[-1, 1] \Rightarrow a=-1, b=1$ into the $t, x$, we found that $x=t$. Hence, we can write
$
\begin{align}
T_k(t) = T_k(x) = \cos(k \arccos(x))
\end{align}
$
Finding the roots for this implies finding the nodes. The nodes are therefore given as:
$
\begin{align}
x_k = \cos\left(\frac{(2k+1)\pi}{2N}\right), \text{where } k=0, 1, ..., N
\end{align}
$
We then have the equation we need in order to implement a function that computes the Chebyshev nodes for a given $N$.

```python
def cheb_nodes(N):
    k_values = np.arange(1, N+1, dtype=float)
    nodes = np.cos( (2 * (k_values) + 1) * np.pi / (2 * N))
    return nodes
```

**b) 2) Implement a funciton "cheb_coef(y,t)" which computes the coefficients $c_k$.**

```python
def cheb_coef(y, t):
    N = len(y)
    coeffs = [(2 / N) * np.sum(y * chebyshev_T(n=k, x=t)) for k in range(N)]
    coeffs[0] = (1/N) * np.sum(y)
    return coeffs
```

**b) 3)Implement a function $\texttt{cheb\_interpol(c,s)}$ which computes and evaluates the interpolant $T_N$ (as a linear combination of the coefficients and the Chebyschev polynomials up to degree $N$) in the test-nodes $s$.**

Use that since we are at the interval $[-1, 1]$, it follows that $t=x$, such that:
$
\begin{equation}
T_k(x_i) = \cos(k*\arccos(x_i))
\end{equation}
$

```python
def cheb_interpol(c, s):
    N = len(c)
    T_N = sum([c[i] * chebyshev_T(n=i, x=s) for i in range(N)])
    return T_N
```

Test the implementation with a dummy function
$
\begin{equation}
f(x) = \cos(10x)
\end{equation}
$

```python
# Define the function you want to interpolate
def f(x):
    return np.cos(10*x)

# Number of Chebyshev coefficients (increase for better precision)
N = 20

# Compute Chebyshev nodes
x_nodes = cheb_nodes(N)
y_nodes = f(x_nodes)

# Compute the Chebyshev coefficients
coeffs = cheb_coef(y=y_nodes, t=x_nodes)

# Reconstruct the function
s_values = np.linspace(-1, 1, 400)
interpol = cheb_interpol(c=coeffs, s=s_values)

# Plot the original function and the Chebyshev approximation
plt.plot(s_values, f(s_values), 'k-', label="f(x)")
plt.plot(s_values, interpol, 'r--', label=f"Chebyshev approximation - N = {N}")
plt.scatter(x_nodes, y_nodes, c='blue', marker='o', label="Chebyshev Nodes")
plt.legend()
plt.grid(True)
plt.title("Chebyshev Polynomial Interpolation of f(x)=cos(10x)")
plt.show()
```
