### **Problem 1:**

**a) Prove that, for any nonsingular matrix $A \in \mathbb{R}^{n \times n}$,
$
\begin{equation} 
\kappa_2 (A) = \left(\frac{\lambda_{max}}{\lambda_{min}}\right)^{1/2},
\end{equation}
$
where $\lambda_{min}$ is the smallest and $\lambda_{max}$ is the largest eigevalue of the matrix $A^TA$.**

From the defintion, we have that the condition number of a matrix $A \in \mathbb{C}^{n \times n}$ is $K(A)_2 = \parallel A\parallel_2  \parallel A^{-1}\parallel_2 $. 

The defintion of the norm-2 of a matrix is $\parallel A\parallel_2 = \sqrt{\lambda_{max}}$.
In order to obtain the expression for the norm-2 of the inverse of A, $\parallel A^{-1}\parallel_2$, I am using SVD to write 
\begin{equation}
A = U\Sigma V^H,
\end{equation}
where $U \in \mathbb{C}^{n \times n}$ and $V \in \mathbb{C}^{n \times n}$ are two unitary matrices and $\Sigma=\text{diag}\left(\sigma_1,...,\sigma_n\right) \in \mathbb{R}^{n \times n}$ is a diagonal matrix with the singular values of $A$ as diagonal entries.

Thus we can write,
$
\begin{align}
A &= U\Sigma V^H \\
\left(A\right)^{-1} &= \left(U \Sigma V^H \right)^{-1} \\
A^{-1} &= \left(V^H\right)^{-1} \Sigma^{-1} U^{-1}
\end{align}
$
As taking the inverse of an unitary matrix results in an unitary matrix, we can identify $\Sigma^{-1}$ as the diagonal matrix holding the singular values of $A^{-1}$ as diagonal entries. 
Look at the inverse of a general diagonal matrix,
$
\begin{equation}
\Sigma = 
    \begin{bmatrix} 
        \sigma_1 & 0 & \cdots & 0  \\
        0 & \sigma_2 & 0 & 0  \\
        \vdots & 0 & \ddots & 0  \\
        0 & 0 & 0 & \sigma_n  \\
    \end{bmatrix} 
\end{equation}
$
then
$
\begin{equation}
\Sigma^{-1} = 
    \begin{bmatrix} 
        \frac{1}{\sigma_1} & 0 & \cdots & 0  \\
        0 & \frac{1}{\sigma_2} & 0 & 0  \\
        \vdots & 0 & \ddots & 0  \\
        0 & 0 & 0 & \frac{1}{\sigma_n} \\
    \end{bmatrix} 
\end{equation}
$

Since we are looking for the largest eigenvalue in order to compute $\parallel A^{-1}\parallel_2$, we can see from the structure of the diagonal elements of $\Sigma^{-1}$ that this would be equivalent with $\frac{1}{\lambda_{min}}$. It then follows that we can write $\parallel A^{-1} \parallel_2 = \sqrt{\frac{1}{\lambda_{min}}}$.

Hence,
$
\begin{equation}
K(A)_2 = \parallel A\parallel_2  \parallel A^{-1}\parallel_2  
= \sqrt{\lambda_{max}} \sqrt{\frac{1}{\lambda_{min}}} 
= \sqrt{\frac{\lambda_{max}}{\lambda_{min}}}.
\end{equation}
$

**b) Show that the condition number $\kappa(Q)$ of an orthogonal matrix Q is equal to 1.**

If Q is an orthogonal matrix, then $Q^{-1} = Q^{T}$. That implies that $Q^{T}Q = Q^{-1}Q = \mathbb{I}$, where $\mathbb{I}$ is the Identity matrix. We therefore know that all eigenvalues of $Q^TQ$ has to be equal.
We can see this through finding the eigenvalue of $Q^TQ$ that corresponds to a eigenvector $\vec{v}$ by $Q^TQ \vec{v} = \lambda \mathbb{I} \vec{v}$. 
Since $Q^TQ=\mathbb{I}$, it follows that we will recieve the same eigenvalue for all eigenvectors, meaning that $\lambda_{min} = \lambda_{max}=\lambda$. From a) we have that $\kappa(Q)_2 = \sqrt{\frac{\lambda_{max}}{\lambda_{min}}}$, thus $\kappa(Q)_2 = \sqrt{\frac{\lambda_{max}}{\lambda_{min}}}=\sqrt{\frac{\lambda}{\lambda}} = 1$

**c) Conversely, if $\kappa_2(A) = 1$ for the matrix A, show that all the eigenvalues of $A^TA$ are equal; deduce that A is a scalar multiple of an orthogonal matrix.**

Starting out from the same expression for $\kappa_2$ as we have proven in a)
$
\begin{align}
\kappa_2 (A) = 
1 &= \left(\frac{\lambda_{max}}{\lambda_{min}}\right)^{1/2} \\
1^2 &= \frac{\lambda_{max}}{\lambda_{min}} \\
\lambda_{min} &= \lambda_{max} 
\end{align}
$
such that by considering a matrix A with condition number $\kappa_2(A)=1$ it follows that the smallest and largest eigenvalue for $A^TA$ has to be equal. For the rest of the problem, we will call this equal eigenvalue for $\lambda$.

Since all the eigenvalues are equal, the only possible eigenvalue problem possible to set up has the form
$
\begin{align}
A^TA\vec{v} = \lambda\mathbb{I}\vec{v},
\end{align}
$
where $\mathbb{I}$ is the identity matrix. We can therefore substitude $A=\sqrt{\lambda}Q$, and consider the eigenvalue problem once more:
$
\begin{align}
\left(\sqrt{\lambda}Q\right)^T\sqrt{\lambda}Q &= \lambda\mathbb{I}\vec{v} \\
\lambda Q^TQ &= \lambda\mathbb{I}\vec{v}.
\end{align}
$
Thus, we can observe that for the right- and left hand side of the equation above to fulfill the equality, $Q^TQ=\mathbb{I}$. Hence $Q^T=Q^{-1}$ which is the defintion of an orthogonal matrix. We have then shown that the matrix $A$ is an scalar mulitple of an orthogonal matrix, $A=\sqrt{\lambda}Q$.

<ol>

    \begin{bmatrix}

**b) Show that $\langle A, B\rangle = tr (AB^T)$ defines an inner product on the vector space $2 \times 2$ real matrices.**

To show that $\langle A, B\rangle = tr (AB^T)$ defines the inner product on the vector space of $2 \times 2$ real matrices, we will see whether the following is satisfied:

1. $\langle A, B\rangle = \langle B, A\rangle$ (symmetry)
2. $\langle cA+B, C\rangle = c\langle A, C\rangle + \langle B, C\rangle$ (linearity)
3. $\langle A, A\rangle > 0$, unless $A=0$ (definiteness)

Start by consider number 1.:
$
\begin{align}
\langle A, B\rangle &= tr(AB^T) \\
&= \sum_{k=1}^2\sum_{i=1}^2 a_{ik}b_{ik} \\
&= \sum_{i=1}^2\sum_{k=1}^2 b_{ik}a_{ik} \\
&= tr(BA^T) \\
&= \langle B, A\rangle
\end{align}
$ 

Condsider number 2.:
$
\begin{align}
\langle cA+B, C\rangle &= tr\left((cA+B)C^T\right) \\
&= tr\left(cAC^T + BC^T\right) \\
&= tr(cAC^T) + tr(BC^T) \\
&= ctr(AC^T) + tr(BC^T) \\
&= c\langle A, C\rangle + \langle B, C\rangle
\end{align}
$ 

Consider number 3.:
$
\begin{align}
\langle A, A\rangle &= tr\left(AA^T\right) \\
&= \sum_{k=1}^2\sum_{i=1}^2 a_{ik}a_{ik} \\
&= \sum_{k=1}^2\sum_{i=1}^2 a_{ik}^2 
\end{align}
$ 
Since the square of a real number is always positive, we have that $\langle A, A\rangle = \sum_{k=1}^2\sum_{i=1}^2 a_{ik}^2 > 0$. However, there is one special case of $A$ where the the square is zero; $a_{ik}=0$, implying $A=0$. It that case, $\langle A, A\rangle = \sum_{k=1}^2\sum_{i=1}^2 a_{ik}^2 = 0$.

Since both $A$ and $B$ is skew-symmatric, we have that
$
\begin{equation}
A=-A^T, B=-B^T
\end{equation}
$
Multiplying $A$ and $B$ gives
$
\begin{align}
AB &= \left(-A^T\right)\left(-B^T\right) \\
&= A^TB^T.
\end{align}
$
Let us call this product for matrix $C$, such that $C=AB$. In order to show whether $C$ is skew-symmatric, we will investigate if we can write $C=-C^T$. If this equality is always true, then the product of two skew-symmatric matrices is always a skew-symmatric matrix.
$
\begin{align}
-C^T &= -\left(AB\right)^T \\
&= -B^TA^T \\
&= -(-B)(-A) \\
&= -BA \neq AB = -C
\end{align}
$
If $C$ is to be skew-symmatric, we would have been able to write $-C^T=C$. This would only be possible if could always write the equality $-BA = AB$. However, this is not the case always. Hence, for the product of two skew-symmatric matrices is not always skew-symmatric.


Next, considering the general form of $A^2$: Given a skew-symmatric matrix $A$, we have that $A=-A^T$. Let us denote a matrix $D=A^2 = (-A^T)(-A^T) = (A^T)^2=(A^2)^T$. I wish to investigate the properties of this matrix by looking at the transpose, $D^T$.

$
\begin{align}
D^T &= \left[(A^2)^T\right]^T = A^2 = D
\end{align}
$
Thus, we can observe that the square of a skew-symmatric matrix $A$, $A^2$, is in general a symmatric matrix.

**This scheme is of order $\mathcal{O}(h^2)$, for what order of $h$ should we see a relevant rounding error?**


With a scheme of order $\mathcal{O}(h^2)$, we have that $u(x+h) - u(x) = u'(x) + Ch^2$. 
Thus, to investigate when the rounding error is comparable with the approximation error, we will need to consider 
$
\begin{align}
Ch^2 &\approx \frac{\epsilon \mid u(x)\mid}{h} \\
h &\approx \left( \frac{\epsilon \mid u(c)\mid}{C} \right)^{1/3} = \mathcal{O}(\epsilon^{1/3})
\end{align}
$
Since the machine accuracy is at $\epsilon=10^{-16}$, we should expect a relevant rounding error when h reaches $\mathcal{O}(\epsilon^{1/3}) \approx 10^{-16/3} \approx 10^{-5.33}$.


```python
import matplotlib.pyplot as plt
import numpy as np

def func_u(x):
    return np.exp(x)

def forward_diff(x, h):
    return (func_u(x+h) - func_u(x)) / (h)

def central_diff(x, h):
    return (func_u(x+h) - func_u(x-h)) / (2*h)

x_0 = 0
h_values = np.logspace(1, -14, int(1e2), endpoint=True)

forward_values = [forward_diff(x_0, h) for h in h_values]
central_values = [central_diff(x_0, h) for h in h_values]
forward_error = [np.abs(f - func_u(x_0)) for f in forward_values]
central_error = [np.abs(c - func_u(x_0)) for c in central_values]

plt.plot(h_values, forward_error, '.-', color="black", label="forward")
plt.plot(h_values, central_error, '.-', color="green", label="central")
plt.xlabel("h")
plt.ylabel("absolute error")
plt.title("Problem 4b")
plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.show()
```

From the plot above, we can observe that our calculations are confirmed. The forward differentiation method change behavior when the $h\sim 10^{-8}$, while central differentiation change behavior when $h \sim 10^{-5}$.

**Central differentiation 2. What happens if we use the central differentiation to approximate a second order derivative of a function? Consider
$
\begin{equation}
u''(x) \approx \frac{u(x-h) -2u(x) + u(x+h)}{h^2}
\end{equation}
$
What is the order of this scheme? For what order of $h$ should we expect to see a relevant rounding error?
To confirm your estimate, use this scheme to differentiate the same function $u(x) = e^x$ as in point b) at $x = 0$. In a plot show the error when using different values of $h$.**

To find the order of the scheme, we compare to the taylor expansion. I can observe that this expression contains both $u(x-h)$ and $u(x+h)$. I will therefore taylor expand both these expression.

$
\begin{align}
u(x+h) &= u(x) + h u'(x) + h^2 \frac{u''(x)}{2!} - h^3 \frac{u^{(3)}(x)}{3!}  +h^4 \frac{u^{(4)}(x)}{4!} + \mathcal{O}(h^5) \\
u(x-h) &= u(x) - h u'(x) + h^2 \frac{u''(x)}{2!} + h^3 \frac{u^{(3)}(x)}{3!}  +h^4 \frac{u^{(4)}(x)}{4!} + \mathcal{O}(h^5) \\
\end{align}
$

Since we do not have any first derivative in our Central Differential 2 expression, we want to get rid of all parts contanig the first derivative. We see that this can be achieved by adding $u(x+h) + u(x-h)$:

$
\begin{align}
u(x+h) + u(x-h) &= 2u(x) + h^2 u''(x) + 2h^4 \frac{u^{(4)}(x)}{4!}+ \mathcal{O}(h^5) \\
\frac{u(x+h) + u(x-h) - 2u(x)}{h^2} &= u''(x) + 2\frac{h^4}{h^2} \frac{u^{(4)}(x)}{4!} + \frac{1}{h^2}\mathcal{O}(h^5) \\
\Rightarrow u''(x) &= \frac{u(x+h) + u(x-h) - 2u(x)}{h^2} + \mathcal{O}(h^2)
\end{align}
$

We recongnice the right hand term as the approximation for a second order derivative function pluss + a term of order $\mathcal{O}(h^2)$. Thus, this approximation scheme is of order $h^2$.

Let us compute the approximation error scheme, to see when we will expect the rounding error to be relevant.

We have the rounded difference as
$
\begin{align}
\tilde\Delta u(x) &= u(x+h) - 2u(x) + u(x-h) \pm \epsilon\mid u(x)\mid
\end{align}
$
Devide by $h^2$ to achieve the formulation for the second differention function
$
\begin{align}
\frac{\tilde\Delta u(x)}{h^2} &= \frac{u(x+h) - 2u(x) + u(x-h)}{h^2} \pm \frac{\epsilon\mid u(x)\mid}{h^2}
\end{align}
$
Since we have shown that the second order differention function is a method of order $\mathcal{O}(h^2)$, we know that
$
\begin{align}
\frac{u(x+h) - 2u(x) + u(x-h)}{h^2} = u''(x) + Ch^2
\end{align}
$
Combinding this gives
$
\begin{align}
\frac{\tilde\Delta u(x)}{h^2} &= u''(x) + Ch^2 \pm \frac{\epsilon\mid u(x)\mid}{h^2} \\
\frac{\tilde\Delta u(x)}{h^2} - u''(x) &=  Ch^2 \pm \frac{\epsilon\mid u(x)\mid}{h^2}
\end{align}
$

So, to see when we would expect the rounding error to be relevant, we will investigate when the two parts on the right side of the equation is equal in absolute value:
$
\begin{align}
\mid Ch^2 \mid \approx \frac{\epsilon\mid u(x)\mid}{h^2} \\
h \approx \left( \frac{\mid u(x)\mid}{C}\epsilon \right)^{1/4} = \mathcal{O}(\epsilon^{1/4})
\end{align}
$
Agian, as in probelm b), since the machine accuracy is at $\epsilon=10^{-16}$, we should expect a relevant rounding error when h reaches $\mathcal{O}(\epsilon^{1/4}) \approx 10^{-16/4} \approx 10^{-4}$.

```python
import matplotlib.pyplot as plt
import numpy as np

def func_u(x):
    return np.exp(x)

def second_diff(x, h):
    return (func_u(x-h) - 2*func_u(x) + func_u(x+h)) / (h**2)

x_0 = 0
h_values = np.logspace(1, -7, int(1e2), endpoint=True)

second_values = [second_diff(x_0, h) for h in h_values]
second_error = [np.abs(f - func_u(x_0)) for f in second_values]

plt.plot(h_values, second_error, '.-', color="black", label="second")
plt.xlabel("h")
plt.ylabel("absolute error")
plt.title("Problem 4c")
plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.show()
```

From the plot above, we can observe that our calculations are almost confirmed. 
The second differentiation method change behavior when the $h\sim 10^{-4}$. However, I do observe that this happes at a $h$ slightly smaller than $h\sim 10^{-4}$.
